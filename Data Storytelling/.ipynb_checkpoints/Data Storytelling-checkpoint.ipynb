### 2 Introduction<a id='2_Introduction'></a>

#### 2.1 Hypothesis<a id='2.1_Hypothesis'></a>

There are more musicians now than there were at any point in history. This can be attributed to things like youtube, soundcloud, and the lessening prices of recording equipment. Record labels looking to sign new artists to expensive contracts need to know whether that artist is going to be worth the investment. Spotify's top ten tracks provide an excellent way to investigate the factors that make a song top the charts. I will use the top ten tracks from Spotify throughout the 2010s to determine the qualities of chart topping songs to guide record labels in deciding which artists will be the most successful based on the attributes of their music. I hypothesis that music that is happier and is easier to dance to will be the music that is most successful.

#### 2.2 Data Sources<a id='2.2_Data_Sources'></a>

The data for this project comes from the a dataset of the top spotify songs from 2010 - 2019. The dataset contains a total of around 600 top ten tracks from across those ten years. The dataset also lists a number of useful features, listed below. <br>
bpm - beats per minute <br>
nrgy - a rating from zero to 100 of how energetic the song is <br>
dnce - a rating from zero to 100 of how easy the song is to dance to <br>
dB - volumne <br>
live - whether the track is a live recording or not <br>
val - a rating from zero to 100 of how positive the song is <br>
dur - the duration of the song <br>
acous - a rating from zero to 100 of how much of the song is acoustic <br>
spch - a rating from zero to 100 of how much of the song is spoken word  <br>
pop - a rating from zero to 100 of how popular the song is <br>

the dataset can be found at: <br>
https://www.kaggle.com/datasets/leonardopena/top-spotify-songs-from-20102019-by-year

### 6 Explore<a id='6_Explore'></a>

```python
Genres = list(topTens['top genre'].unique())
Artists = list(topTens['artist'].unique())
```

Let's start off by looking at a correlation heatmap for our dataset. From this we can see how our different features effect one another

```python
plt.subplots(figsize = (12, 8))
sns.heatmap(topTens.corr(), annot = True);
```

Many of the variables have pretty low correlation to eachother. There are a few higher correlations, though. Dancibility and positivity seem to be highly correlated with one another, likewise energy and volume are highly correlated. Let's take a closer look at some of these relationships

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 8), squeeze=False)
fig.subplots_adjust(wspace=0.5, hspace=0.6)
fig.suptitle("Relationships between Danceability:Positivity, Volume:Energy")
ax[0,0].scatter(x = 'val', y = 'dnce', data=topTens, alpha=0.5)
ax[0,0].set(xlabel='Positivity', ylabel='Danceability')
ax[0,1].scatter(x = 'nrgy', y = 'dB', data=topTens, alpha=0.5)
ax[0,1].set(xlabel='Energy', ylabel='Volume')
plt.show()
```

Interesting! High positivity songs are much easier to dance to. This is not too surprising. Let's look at the distributions of our different variables and see what that tells us

```python
topTens.hist(figsize = (15, 10))
plt.subplots_adjust(hspace=0.8, wspace = 0.5);
```

High energy songs seem to be the leader, no question. Danceability also seems to be a really important factor in whether a song will be a chart topper. Another interesting thing we see from the distributions is that songs longer than four minutes have a tough time being in the top ten. This could also be that there are far fewer songs that are this long, but it is still interesting to see

Spoken words seem to also be a hinderance to success, as well as acoustic sections. This makes sense. The correlation matrix tells us that acoustic songs are inversely correlated to energy, which we know is an important factor.

Let's look at these factors by year. Then, we can see if opinions are changing or if they are consistent throughout the decade

```python
fig, ax = plt.subplots(3, 2, figsize=(15, 8), squeeze=False)
fig.subplots_adjust(wspace=0.5, hspace=0.6)
fig.suptitle("Feature averages by year")
ax[0,0].plot(topTens['year'].unique(), topTens.groupby('year')['dnce'].mean())
ax[0,0].set(xlabel='Year', ylabel='Average Danceability')
ax[0,1].plot(topTens['year'].unique(), topTens.groupby('year')['val'].mean())
ax[0,1].set(xlabel='Year', ylabel='Average Positivity')
ax[1,0].plot(topTens['year'].unique(), topTens.groupby('year')['nrgy'].mean())
ax[1,0].set(xlabel='Year', ylabel='Average Energy Level')
ax[1,1].plot(topTens['year'].unique(), topTens.groupby('year')['dur'].mean())
ax[1,1].set(xlabel='Year', ylabel='Average Duration')
ax[2,0].plot(topTens['year'].unique(), topTens.groupby('year')['bpm'].mean())
ax[2,0].set(xlabel='Year', ylabel='Average Beats per Minute')
ax[2,1].plot(topTens['year'].unique(), topTens.groupby('year')['spch'].mean())
ax[2,1].set(xlabel='Year', ylabel='Average Spoken Word Ratio')
plt.show()
```

The values for these factors are all pretty close, meaning opinions are pretty consistent throughout the years. We do see, however, a slight increase in the favorability of dancier tracks towards the end of the 2010s. Although spoken words are not seen too favorably, the amount that they show up is never zero. While danceability is up, energy and beats per minute are all down. This is really interesting as it does not align with the correlations that we saw earlier

Finally, let's look at genres and artists.

```python
topTenArtists = topTens.groupby('artist')['title'].count().sort_values(ascending = False)[:10]
topTenGenres = topTens.groupby('top genre')['title'].count().sort_values(ascending = False)[:10]
```

```python
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(8)
plt.bar(topTenArtists.index, topTenArtists.values)
plt.xlabel('Artist')
plt.ylabel('Number of top ten tracks')
plt.title('Number of top 10 tracks by artist')
plt.xticks(rotation = 45)
plt.show()
```

Katy Perry, Justin Bieber, and Maroon 5 are the top three artists. Let's look more at them specifically.

```python
KP = topTens[topTens['artist'] == 'Katy Perry']
JB = topTens[topTens['artist'] == 'Justin Bieber']
M5 = topTens[topTens['artist'] == 'Maroon 5']
```

```python
print(KP['top genre'].unique())
print(JB['top genre'].unique())
print(M5['top genre'].unique())
```

Wow! They all make pop music. Not only that, but they seem to only make that one genre of music. Perhaps being an artist that specializes in a particular type of music helps with consistent success

Let's look at the energy levels of the top ten artists

```python
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(8)
plt.bar(topTenArtists.index, topTens[topTens['artist'].isin(topTenArtists.index)].groupby('artist')['nrgy'].mean())
plt.xlabel('Artist')
plt.ylabel('Number of top ten tracks')
plt.title('Number of top 10 tracks by artist')
plt.xticks(rotation = 45)
plt.show()
```

Most are fairly high, around 80; in the upper range of the distribution that we saw earlier. However, it is notable to see that both maroon 5 and Justin Bieber are much lower at around 60.

Let's move on to genres

```python
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(8)
plt.bar(topTenGenres.index, topTenGenres.values)
plt.xlabel('Genre')
plt.ylabel('Number of top ten tracks')
plt.title('Number of top 10 tracks by genre')
plt.xticks(rotation = 45)
plt.show()
```

Unsurprisingly, pop music wins out, with dance pop being by far the most popular.

Let's look at genre by year like we did above

```python
topTens.groupby('year')['top genre'].max()
```

Almost every year has pop being the most popular,except for two. 2015 and 2016 had tropical house on top.
